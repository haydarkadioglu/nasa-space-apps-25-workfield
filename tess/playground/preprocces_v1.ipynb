## Data Cleaning
### Identify and Remove Problematic Columns
We'll identify columns with:
1. High percentage of missing values (>50%)
2. Very low variance (imbalanced columns with >95% same value)

```python
# Store original shape
original_shape = df.shape
print(f"Original dataset shape: {original_shape[0]} rows, {original_shape[1]} columns")
print("="*70)
```

```python
# 1. Identify columns with high missing values (>=25%)
missing_threshold = 25  # percentage
missing_percent = (df.isnull().sum() / len(df)) * 100
high_missing_cols = missing_percent[missing_percent > missing_threshold].sort_values(ascending=False)

print(f"\n1. Columns with more than {missing_threshold}% missing values:")
print("-" * 70)
if len(high_missing_cols) > 0:
    for col, pct in high_missing_cols.items():
        print(f"  {col}: {pct:.2f}% missing")
else:
    print(f"  No columns found with >{missing_threshold}% missing values")
```

```python
# 2. Identify imbalanced columns (columns where one value dominates >95%)
imbalance_threshold = 95  # percentage
imbalanced_cols = []

print(f"\n2. Imbalanced columns (>={imbalance_threshold}% of rows have the same value):")
print("-" * 70)

for col in df.columns:
    # Calculate the percentage of the most common value
    if len(df[col]) > 0:
        value_counts = df[col].value_counts(dropna=False)
        if len(value_counts) > 0:
            most_common_pct = (value_counts.iloc[0] / len(df)) * 100
            if most_common_pct >= imbalance_threshold:
                imbalanced_cols.append(col)
                print(f"  {col}: {most_common_pct:.2f}% have value '{value_counts.index[0]}'")

if len(imbalanced_cols) == 0:
    print(f"  No columns found with >={imbalance_threshold}% imbalance")
```

```python
# 3. Combine all columns to remove
cols_to_remove = list(set(high_missing_cols.index.tolist() + imbalanced_cols))

print(f"\n3. Summary of columns to remove:")
print("=" * 70)
print(f"Total columns to remove: {len(cols_to_remove)}")
if len(cols_to_remove) > 0:
    print("\nColumns that will be removed:")
    for col in sorted(cols_to_remove):
        reasons = []
        if col in high_missing_cols.index:
            reasons.append(f"High missing: {high_missing_cols[col]:.2f}%")
        if col in imbalanced_cols:
            reasons.append(f"Imbalanced: >={imbalance_threshold}%")
        print(f"  - {col}: {', '.join(reasons)}")
else:
    print("  No columns need to be removed!")
```

```python
# 4. Create cleaned dataset by removing identified columns
if len(cols_to_remove) > 0:
    df_cleaned = df.drop(columns=cols_to_remove)
    print(f"\n4. Dataset after cleaning:")
    print("=" * 70)
    print(f"Original shape: {original_shape[0]} rows  {original_shape[1]} columns")
    print(f"Cleaned shape:  {df_cleaned.shape[0]} rows  {df_cleaned.shape[1]} columns")
    print(f"Columns removed: {len(cols_to_remove)}")
    print(f"Columns retained: {df_cleaned.shape[1]}")
else:
    df_cleaned = df.copy()
    print("\n4. No columns were removed. Dataset remains unchanged.")
```

```python
# 5. Display remaining columns and their characteristics
print("\n5. Remaining columns after cleaning:")
print("=" * 70)
remaining_missing = (df_cleaned.isnull().sum() / len(df_cleaned)) * 100

for col in df_cleaned.columns:
    missing_pct = remaining_missing[col]
    dtype = df_cleaned[col].dtype
    unique_vals = df_cleaned[col].nunique()
    print(f"  {col}:")
    print(f"    - Type: {dtype}, Unique values: {unique_vals}, Missing: {missing_pct:.2f}%")
```

## Advanced Missing Value Imputation
### Strategy: Group-based Imputation using Planet Characteristics
We'll classify planets based on similar characteristics (pl_rade, pl_eqt, pl_insol) and fill missing values within each group using mean/median.

```python
# Check the current state of key columns
key_columns = ['pl_rade', 'pl_eqt', 'pl_insol']

print("Current state of key planet characteristic columns:")
print("=" * 70)
for col in key_columns:
    if col in df_cleaned.columns:
        missing_count = df_cleaned[col].isnull().sum()
        missing_pct = (missing_count / len(df_cleaned)) * 100
        print(f"\n{col}:")
        print(f"  Missing: {missing_count} ({missing_pct:.2f}%)")
        print(f"  Min: {df_cleaned[col].min():.4f}" if not df_cleaned[col].isnull().all() else "  Min: N/A")
        print(f"  Max: {df_cleaned[col].max():.4f}" if not df_cleaned[col].isnull().all() else "  Max: N/A")
        print(f"  Mean: {df_cleaned[col].mean():.4f}" if not df_cleaned[col].isnull().all() else "  Mean: N/A")
        print(f"  Median: {df_cleaned[col].median():.4f}" if not df_cleaned[col].isnull().all() else "  Median: N/A")
    else:
        print(f"\n{col}: NOT FOUND in cleaned dataset")
```

```python
# Step 1: First, fill the key columns (pl_rade, pl_eqt, pl_insol) with median
# This allows us to use them for clustering
print("\nStep 1: Initial imputation of key columns with median")
print("=" * 70)

df_imputed = df_cleaned.copy()

for col in key_columns:
    if col in df_imputed.columns:
        initial_missing = df_imputed[col].isnull().sum()
        if initial_missing > 0:
            median_value = df_imputed[col].median()
            df_imputed[col].fillna(median_value, inplace=True)
            print(f"{col}: Filled {initial_missing} missing values with median ({median_value:.4f})")
```

```python
# Step 2: Normalize the key features for clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

print("\nStep 2: Clustering planets based on characteristics")
print("=" * 70)

# Prepare features for clustering
clustering_features = []
for col in key_columns:
    if col in df_imputed.columns:
        clustering_features.append(col)

if len(clustering_features) > 0:
    # Standardize the features
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(df_imputed[clustering_features])
    
    # Determine optimal number of clusters using elbow method
    inertias = []
    K_range = range(2, min(11, len(df_imputed)//10 + 2))
    
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(features_scaled)
        inertias.append(kmeans.inertia_)
    
    # Plot elbow curve
    plt.figure(figsize=(10, 5))
    plt.plot(K_range, inertias, 'bo-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal k')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    print(f"Tested cluster sizes: {list(K_range)}")
else:
    print("No clustering features available!")
```

```python
# Step 3: Apply K-Means clustering with optimal k (let's use k=5 as a reasonable choice)
optimal_k = 5

print(f"\nStep 3: Applying K-Means with k={optimal_k}")
print("=" * 70)

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_imputed['planet_cluster'] = kmeans.fit_predict(features_scaled)

# Display cluster distribution
cluster_counts = df_imputed['planet_cluster'].value_counts().sort_index()
print(f"\nCluster distribution:")
for cluster_id, count in cluster_counts.items():
    print(f"  Cluster {cluster_id}: {count} planets ({count/len(df_imputed)*100:.2f}%)")

# Display cluster characteristics
print(f"\nCluster characteristics based on key features:")
cluster_stats = df_imputed.groupby('planet_cluster')[clustering_features].agg(['mean', 'median', 'std', 'count'])
print(cluster_stats)
```

```python
# Step 4: Visualize the clusters in 2D using PCA
from sklearn.decomposition import PCA

print("\nStep 4: Visualizing planet clusters")
print("=" * 70)

# Apply PCA to reduce to 2D for visualization
pca = PCA(n_components=2)
features_2d = pca.fit_transform(features_scaled)

# Create scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(features_2d[:, 0], features_2d[:, 1], 
                     c=df_imputed['planet_cluster'], cmap='viridis', 
                     alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
plt.colorbar(scatter, label='Cluster')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)')
plt.title('Planet Clusters based on Physical Characteristics (PCA visualization)')
plt.grid(True, alpha=0.3)
plt.show()

print(f"Total variance explained by 2 components: {sum(pca.explained_variance_ratio_)*100:.2f}%")
```

```python
# Step 5: Fill missing values within each cluster using group statistics
print("\nStep 5: Filling missing values within each cluster")
print("=" * 70)

# Create a copy with original missing values restored for the imputation process
df_final = df_cleaned.copy()
df_final['planet_cluster'] = df_imputed['planet_cluster']

# Get all numerical columns
numerical_cols_to_fill = df_final.select_dtypes(include=[np.number]).columns.tolist()
numerical_cols_to_fill = [col for col in numerical_cols_to_fill if col != 'planet_cluster']

print(f"Total numerical columns to process: {len(numerical_cols_to_fill)}\n")

# Track imputation statistics
imputation_summary = []

for col in numerical_cols_to_fill:
    initial_missing = df_final[col].isnull().sum()
    
    if initial_missing > 0:
        filled_count = 0
        
        # Fill within each cluster
        for cluster_id in df_final['planet_cluster'].unique():
            cluster_mask = df_final['planet_cluster'] == cluster_id
            cluster_data = df_final.loc[cluster_mask, col]
            
            # Get missing values in this cluster
            missing_in_cluster = cluster_data.isnull().sum()
            
            if missing_in_cluster > 0 and cluster_data.notna().sum() > 0:
                # Use median for imputation (more robust to outliers)
                fill_value = cluster_data.median()
                df_final.loc[cluster_mask & df_final[col].isnull(), col] = fill_value
                filled_count += missing_in_cluster
        
        # If any values are still missing (entire cluster was NaN), use global median
        remaining_missing = df_final[col].isnull().sum()
        if remaining_missing > 0:
            global_median = df_final[col].median()
            df_final[col].fillna(global_median, inplace=True)
            filled_count += remaining_missing
        
        imputation_summary.append({
            'Column': col,
            'Initial Missing': initial_missing,
            'Filled': filled_count,
            'Method': 'Cluster-based median'
        })

# Display imputation summary
if len(imputation_summary) > 0:
    imputation_df = pd.DataFrame(imputation_summary)
    print("Imputation Summary:")
    print(imputation_df.to_string(index=False))
    print(f"\nTotal values imputed: {imputation_df['Filled'].sum()}")
else:
    print("No missing values were found to impute!")
```

```python
# Step 6: Verify the imputation results
print("\nStep 6: Verification of imputation")
print("=" * 70)

# Check for remaining missing values
remaining_missing = df_final.isnull().sum()
remaining_missing = remaining_missing[remaining_missing > 0]

if len(remaining_missing) > 0:
    print("WARNING: Some columns still have missing values:")
    for col, count in remaining_missing.items():
        pct = (count / len(df_final)) * 100
        print(f"  {col}: {count} ({pct:.2f}%)")
else:
    print(" All numerical columns have been successfully imputed!")
    print(f" Dataset now has {len(df_final)} rows with no missing values in numerical columns")

# Remove the temporary cluster column
df_final = df_final.drop(columns=['planet_cluster'])

print(f"\nFinal dataset shape: {df_final.shape[0]} rows  {df_final.shape[1]} columns")
```

```python
# Step 7: Compare before and after imputation
print("\nStep 7: Comparison of missing values before and after imputation")
print("=" * 70)

# Calculate missing percentages
missing_cleaned = (df_cleaned.select_dtypes(include=[np.number]).isnull().sum() / len(df_cleaned)) * 100
missing_final = (df_final.select_dtypes(include=[np.number]).isnull().sum() / len(df_final)) * 100

comparison_df = pd.DataFrame({
    'Column': missing_cleaned.index,
    'Before (%)': missing_cleaned.values,
    'After (%)': missing_final.values,
    'Improvement (%)': missing_cleaned.values - missing_final.values
})

# Show only columns that had missing values
comparison_df = comparison_df[comparison_df['Before (%)'] > 0].sort_values('Before (%)', ascending=False)

if len(comparison_df) > 0:
    print(comparison_df.to_string(index=False))
else:
    print("No missing values were present in the dataset.")
```

## Feature Engineering
Create new features from existing data to improve model performance

```python
# Define function to classify stars by spectral class based on effective temperature
def spectral_class(teff):
    """
    Classify star into spectral class based on effective temperature (Kelvin)
    O: > 30000K (hottest, blue)
    B: 10000-30000K (blue-white)
    A: 7500-10000K (white)
    F: 6000-7500K (yellow-white)
    G: 5200-6000K (yellow, like our Sun)
    K: 3700-5200K (orange)
    M: < 3700K (coolest, red)
    """
    if pd.isna(teff):
        return 'Unknown'
    elif teff > 30000:
        return 'O'
    elif teff > 10000:
        return 'B'
    elif teff > 7500:
        return 'A'
    elif teff > 6000:
        return 'F'
    elif teff > 5200:
        return 'G'
    elif teff > 3700:
        return 'K'
    else:
        return 'M'

print("Spectral Classification Function defined")
print("=" * 70)
print("Spectral Classes:")
print("  O: > 30000K (hottest, blue stars)")
print("  B: 10000-30000K (blue-white stars)")
print("  A: 7500-10000K (white stars)")
print("  F: 6000-7500K (yellow-white stars)")
print("  G: 5200-6000K (yellow stars, like our Sun)")
print("  K: 3700-5200K (orange stars)")
print("  M: < 3700K (coolest, red stars)")
```

```python
# Define function to classify planets by type based on radius
def planet_type(rade):
    """
    Classify planet into type based on radius (Earth radii)
    rocky: < 1.5 Earth radii (Earth-like, rocky planets)
    sub-neptune: 1.5-4 Earth radii (mini-Neptunes)
    gas giant: > 4 Earth radii (Jupiter-like, gas giants)
    """
    if pd.isna(rade):
        return 'Unknown'
    elif rade < 1.5:
        return 'rocky'
    elif rade < 4:
        return 'sub-neptune'
    else:
        return 'gas giant'

print("\nPlanet Classification Function defined")
print("=" * 70)
print("Planet Types:")
print("  rocky: < 1.5 Earth radii (Earth-like, rocky planets)")
print("  sub-neptune: 1.5-4 Earth radii (mini-Neptunes)")
print("  gas giant: > 4 Earth radii (Jupiter-like, gas giants)")
```

## Label Encoding for Categorical Variables
Convert all categorical (string) variables to numerical format for machine learning models

```python
# Identify all categorical (object type) columns
categorical_cols = df_final.select_dtypes(include=['object']).columns.tolist()

print("Categorical Columns Found:")
print("=" * 70)
if len(categorical_cols) > 0:
    for i, col in enumerate(categorical_cols, 1):
        unique_vals = df_final[col].nunique()
        print(f"{i}. {col:25} - {unique_vals} unique values")
        if unique_vals <= 10:
            print(f"   Values: {df_final[col].unique()}")
    print(f"\nTotal categorical columns: {len(categorical_cols)}")
else:
    print("No categorical columns found in the dataset")
```

```python
# Create a dictionary to store all encoding mappings
encoding_mappings = {}

print("\nApplying Label Encoding to Categorical Columns:")
print("=" * 70)

if len(categorical_cols) > 0:
    from sklearn.preprocessing import LabelEncoder
    
    # Create a copy to preserve original data
    df_encoded = df_final.copy()
    
    for col in categorical_cols:
        print(f"\nEncoding column: {col}")
        
        # Initialize LabelEncoder
        le = LabelEncoder()
        
        # Fit and transform the column
        df_encoded[col + '_encoded'] = le.fit_transform(df_final[col])
        
        # Create mapping dictionary
        mapping = dict(zip(le.classes_, le.transform(le.classes_)))
        encoding_mappings[col] = mapping
        
        # Display the mapping
        print(f"  Mapping created: {len(mapping)} unique values")
        for original, encoded in mapping.items():
            print(f"    '{original}' -> {encoded}")
        
        # Drop the original string column
        df_encoded = df_encoded.drop(columns=[col])
        
        # Rename encoded column to original name
        df_encoded = df_encoded.rename(columns={col + '_encoded': col})
        
        print(f"   Column '{col}' successfully encoded")
    
    # Update df_final with encoded version
    df_final = df_encoded.copy()
    
    print(f"\n{'='*70}")
    print(f" All {len(categorical_cols)} categorical columns have been encoded")
    print(f" All encoding mappings have been saved")
else:
    print("No categorical columns to encode")
```

```python
import json

# Display the complete encoding mappings dictionary
print("\nComplete Encoding Mappings Dictionary:")
print("=" * 70)

# Convert numpy int32 values to regular Python int for JSON serialization
encoding_mappings_json = {}
for col, mapping in encoding_mappings.items():
    encoding_mappings_json[col] = {k: int(v) for k, v in mapping.items()}

print(json.dumps(encoding_mappings_json, indent=2))

# Save to a JSON file for future reference
with open('encoding_mappings.json', 'w') as f:
    json.dump(encoding_mappings_json, f, indent=2)

print(f"\n Encoding mappings saved to 'encoding_mappings.json'")
```

```python
# Verify that all columns are now numerical
print("\nVerification: Data Types After Encoding")
print("=" * 70)

numerical_count = len(df_final.select_dtypes(include=[np.number]).columns)
categorical_count = len(df_final.select_dtypes(include=['object']).columns)

print(f"Numerical columns: {numerical_count}")
print(f"Categorical columns: {categorical_count}")

if categorical_count == 0:
    print("\n SUCCESS: All columns are now numerical!")
    print(" Dataset is ready for machine learning models")
else:
    print(f"\n WARNING: Still {categorical_count} categorical columns remaining:")
    remaining_categorical = df_final.select_dtypes(include=['object']).columns.tolist()
    for col in remaining_categorical:
        print(f"  - {col}")

print(f"\nFinal dataset shape: {df_final.shape[0]} rows  {df_final.shape[1]} columns")
```

```python
# Create a reverse mapping function for interpretation
def decode_value(column_name, encoded_value):
    """
    Decode an encoded value back to its original string representation
    
    Parameters:
    - column_name: Name of the encoded column
    - encoded_value: The numerical encoded value
    
    Returns:
    - Original string value
    """
    if column_name in encoding_mappings:
        # Create reverse mapping
        reverse_mapping = {v: k for k, v in encoding_mappings[column_name].items()}
        return reverse_mapping.get(encoded_value, "Unknown")
    else:
        return "Column not found in encoding mappings"

print("\nDecode Function Created:")
print("=" * 70)
print("Use decode_value(column_name, encoded_value) to get original values")
print("\nExample usage:")
if len(encoding_mappings) > 0:
    # Get first encoded column
    first_col = list(encoding_mappings.keys())[0]
    first_encoded_val = list(encoding_mappings[first_col].values())[0]
    decoded = decode_value(first_col, first_encoded_val)
    print(f"  decode_value('{first_col}', {first_encoded_val}) = '{decoded}'")
```

```python
# Summary of encoding process
print("\nEncoding Process Summary:")
print("=" * 70)
print(f"Total columns before encoding: {len(df_final.columns)}")
print(f"Categorical columns encoded: {len(encoding_mappings)}")
print(f"Final numerical columns: {len(df_final.select_dtypes(include=[np.number]).columns)}")
print(f"\nEncoded columns:")
for col in encoding_mappings.keys():
    print(f"   {col}: {len(encoding_mappings[col])} categories")

print(f"\n{'='*70}")
print("Dataset is now fully prepared for machine learning!")
print(f"  - All values are numerical")
print(f"  - Encoding mappings saved")
print(f"  - Ready for model training")
```
